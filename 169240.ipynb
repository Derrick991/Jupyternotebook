```python

##Getting the account balance 
def main():
    balance = float(input("Enter your initial account balance: "))
##Display the menu
    while True:
        print("Menu:\n")
        print("1. Withdraw")
        print("2. Deposit")
        print("3. Exit")
##Input of choice
        choice = input("Enter your choice: ")
##Checking the input
        if choice == '1':
            balance = withdraw(balance)
##Print new updated balance
            print("The new balance", balance)
        elif choice == '2':
            balance = deposit(balance)
            print("The new balance", balance)
        elif choice == '3':
            print("Exiting program.")
            break
##Input is not according to the choices            
        else:
            print("Invalid choice. Please try again.")
##Withdrawing money from the current account balance
def withdraw(balance):
    amount = float(input("Enter the amount to withdraw: "))
##Check if the amount to be withdrawn is applicable
    if amount > balance:
        print("Insufficient balance.")
    else:
        balance -= amount
        print("Withdrawal successful.")
    return balance
##Adding a deposit to the current acount balance
def deposit(balance):
    amount = float(input("Enter the amount to deposit: "))
    balance += amount
    print("Deposit successful.")
    return balance

##Write transaction to file
    with open("transactions.txt", "a") as file:
        file.write(f"Choice: {choice}, Balance: {balance}\n")

if __name__ == "__main__":
    main()
```

```python
##importing the necessary library
from random import randint

def guessing_game():
    # Random numbers between 1 to 100
    num = randint(1, 100)

    # Set chances to 5
    chances = 5

    # loop the chances
    while chances > 0:
        # Get the input a random number and store in the varaible guess
        guess = int(input("Guess the number between 1 and 100: "))

        # Check if the guess is correct
        if guess == num:
            print("Congratulations!.")
            return

        # Check if the guess is too low
        if guess < num:
            print("Too low.")

        # Check if the guess is too high
        if guess > num:
            print("Too high.")

        # Decrease the chances by 1
        chances -= 1

    # Exhaustion of chances
    print("Game Over.")

# Run the game
guessing_game()
```
